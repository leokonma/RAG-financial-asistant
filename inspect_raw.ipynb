```python
import pandas as pd

print("üìï Loading SD raw...")

sd_raw = pd.read_excel("data/SD_Transaccions.xlsx",skiprows=6, header=0)

print("Shape:", sd_raw.shape)
sd_raw.head(30)
```

```python
sd = sd_raw.copy()

sd = sd.drop(columns=["Saldo", "Fecha valor"], errors="ignore")
sd.head()
```

```python
import re

def clean_amount_raw(x):
    if pd.isna(x):
        return None
    x = str(x)
    x = x.replace("EUR", "").replace("‚Ç¨", "")
    x = x.replace("‚àí", "-")     # unicode minus ‚Üí standard minus
    x = x.replace(",", ".")     # comma ‚Üí dot
    x = x.strip()

    match = re.findall(r"-?\d+\.?\d*", x)
    return float(match[0]) if match else None

sd["Importe"] = sd["Importe"].apply(clean_amount_raw)
sd["Importe"] = pd.to_numeric(sd["Importe"], errors="coerce")

sd["Importe"].head()
```

```python
sd["Fecha operaci√≥n"] = pd.to_datetime(
    sd["Fecha operaci√≥n"],
    dayfirst=True,
    errors="coerce"
)
sd["Fecha operaci√≥n"].head()
```

```python
sd["Concepto"] = sd["Concepto"].astype(str).str.strip()
sd["Concepto"].head()
```

```python
sd["source"] = "SD"
```

```python
sd["Divisa"] = sd["Divisa"].astype(str).str.strip()
```
